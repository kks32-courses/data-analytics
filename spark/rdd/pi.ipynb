```pyspark3
# pick random points in the unit square ((0, 0) to (1,1)) and 
# see how many fall in the unit circle
# This should be Pi / 4.
def inside(p):
    x, y = random.random(), random.random()
    return x*x + y*y < 1

# Number of samples
NUM_SAMPLES = 100

# Count the number of points in the circle
count = sc.parallelize(range(0, NUM_SAMPLES)) \
             .filter(inside).count()
    
# Estimate Pi
print ("Pi is roughly %f", (4.0 * count / NUM_SAMPLES))
```
